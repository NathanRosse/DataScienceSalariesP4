```python
# Import dependancies
import pandas as pd
from pathlib import Path
```

```python
# Import raw data
df_orig = pd.read_csv(Path("../Data/Latest_Data_Science_Salaries.csv"))
df_orig
```

```python
# Drop irrelevant fields
# Expertise level does not add any value beyond what is in Experience Level.
# Since we are going to filter to only United States the Salary Currency and Salary in USD fields don't add info

df = df_orig.drop(columns=['Expertise Level', 'Salary Currency', 'Salary in USD'])
df
```

```python
# Filter to only full-time employees living in United States

df = df[(df['Employee Residence'] == 'United States') & (df['Employment Type'] == 'Full-Time')]
df
```

```python
# Drop fields for Employment Type and Employee Residence now that they have been filted to only one value.

clean_df = df.drop(columns=['Employment Type', 'Employee Residence'])
clean_df
```

```python
# Export to csv file

clean_df.to_csv('../Data/Clean_Data_Science_Salaries.csv', index=False)
```
